```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import pytz

# Supongamos que tienes un DataFrame llamado datos con una columna 'published_at'
# Convierte la columna 'published_at' en objetos datetime
datos['published_at'] = pd.to_datetime(datos['published_at'], utc=True)

# Elimina la información de zona horaria
datos['published_at'] = datos['published_at'].dt.tz_convert(None)

# Extrae el año de las fechas y crea una nueva columna 'year'
datos['year'] = datos['published_at'].dt.year

# Cuenta las publicaciones por año
publicaciones_por_año = datos['year'].value_counts().sort_index()

# Crea un gráfico de barras para visualizar la distribución de publicaciones por año
plt.figure(figsize=(10, 5))
sns.barplot(x=publicaciones_por_año.index, y=publicaciones_por_año.values)
plt.xlabel('Año de Publicación')
plt.ylabel('Número de Publicaciones')
plt.title('Distribución de Publicaciones por Año')
plt.xticks(rotation=45)
plt.show()
```

En este gráfico podemos apreciar que el año que más publicaciones tuvo fue el 2017. El 2018 y 2020 estuvieron muy parejos pero no pasaron de las 1000

A continuación veremos un diagrama de barras qwue nos muestra cuantos ganadores hay de en estos años de las categorias que desde mi punto de vista son las más importantes

```python
import pandas as pd
import matplotlib.pyplot as plt

categorias_seleccionadas = ["Record Of The Year", "Album Of The Year", "Song Of The Year"]
filtro_categorias = datos['category'].isin(categorias_seleccionadas)
datos_filtrados = datos[filtro_categorias]

conteo_categorias = datos_filtrados['category'].value_counts()

plt.figure(figsize=(10, 6))
conteo_categorias.plot(kind='bar')
plt.title('Recuento de Ganadores por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Recuento de Ganadores')
plt.show()
```

En esta gráfica podemos ver que en comparación entre estos artistas todos tienen por lo menos un premio, pero hay 2 artistas que ganaron 3

```python
import pandas as pd
import matplotlib.pyplot as plt
artistas_interes = ["Wisin y Yandel", "José Feliciano", "Skrillex", "Chris Botti", "Kelly Clarkson"]

filtro_artistas = datos['artist'].isin(artistas_interes)
datos_filtrados = datos[filtro_artistas]

conteo_artistas = datos_filtrados['artist'].value_counts()

plt.figure(figsize=(10, 6))
conteo_artistas.plot(kind='bar')
plt.title('Recuento de Ganadores por Artista')
plt.xlabel('Artista')
plt.ylabel('Recuento de Ganadores')
plt.show()
```
