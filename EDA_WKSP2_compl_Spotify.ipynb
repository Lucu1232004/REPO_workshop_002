```python
import pandas as pd

def leer_csv(ruta_archivo):
    try:
        df = pd.read_csv(ruta_archivo)
        df.dropna(inplace=True)
        return df
    except FileNotFoundError:
        print("Error: El archivo CSV no se encuentra en la ubicación especificada.")
        return None

def transformar_dataframe(df):
    # Función para eliminar todo después de la primera coma
    def eliminar_despues_de_coma(entrada):
        return entrada.split(',')[0]

    # Función para eliminar todo después del primer punto y coma
    def eliminar_despues_de_punto_y_coma(entrada):
        return entrada.split(';')[0]

    # Aplicar las funciones de limpieza a la columna 'artists'
    df['artists'] = df['artists'].apply(eliminar_despues_de_coma)
    df['artists'] = df['artists'].apply(eliminar_despues_de_punto_y_coma)

    # Eliminar columnas no deseadas
    columnas_a_eliminar = ['Unnamed: 0', 'explicit', 'danceability', 'energy', 'loudness', 'danceability',
                            'mode', 'speechiness', 'acousticness', 'instrumentalness', 'valence',
                           'time_signature', 'track_id', 'key']
    df = df.drop(columnas_a_eliminar, axis=1)

    return df

# Ruta al archivo CSV en Google Drive
ruta_del_archivo = "/content/spotify_dataset.csv"

# Llamada a la función leer_csv para obtener el DataFrame
df_original = leer_csv(ruta_del_archivo)

if df_original is not None:
    # Llamada a la función transformar_dataframe y asignación a un nuevo DataFrame
    df_transformado = transformar_dataframe(df_original)

    # Imprimir el DataFrame resultante
    print(df_transformado)
```

Covarianza:
Esta nos permite conocer cómo se comportan las variables en cuestión respecto de otras variables.

Si la covarianza es positiva, significa que cuando una variable tiende a aumentar, la otra también tiende a aumentar.
Si la covarianza es negativa, significa que cuando una variable tiende a aumentar, la otra tiende a disminuir.
Si la covarianza es cercana a cero, no hay una relación lineal fuerte entre las dos variables.

```python
datos.cov()
```

La correlación mide la fuerza y la dirección de la relación lineal entre dos variables

El valor de correlación oscila entre -1 y 1.
Un valor de correlación de 1 indica una correlación positiva perfecta, lo que significa que a medida que una variable aumenta, la otra también aumenta en una relación lineal perfecta.
Un valor de correlación de -1 indica una correlación negativa perfecta, lo que significa que a medida que una variable aumenta, la otra disminuye en una relación lineal perfecta.
Un valor de correlación cercano a 0 indica que no hay una relación lineal fuerte entre las dos variables.

```python
datos.corr()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = datos['duration_ms'].tolist()

fig, ax = plt.subplots()
datos['duration_sec'] = datos['duration_ms'] / 1000
boxprops = dict(linewidth=2, color='blue', linestyle='--')
whiskerprops = dict(linewidth=2, color='red', linestyle='-.')

bp = ax.boxplot(data, vert=False, patch_artist=True, boxprops=boxprops, whiskerprops=whiskerprops)
colors = ['lightblue', 'lightgreen', 'lightpink']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

ax.set_yticklabels(['Duración'])

ax.set_title("Diagrama de Caja de Duración")

plt.show()
```

Vemos que en el gráfico se puede interpretar la duración (0, 1, 2, 3, 4, 5) como en 100000 ya que al estar en milisegundos le quitamos los ceros para que quede mejor en la grafica, pasandolo a segundos quedaria en valores de 100, 200, 300, 400, 500. Por eso vemos muchas canciones de 100 y 200 segundos ya que son el tiempo promedio que maneja una canción normal, pero hay algunas que duran mucho más que el promedio o menos.
